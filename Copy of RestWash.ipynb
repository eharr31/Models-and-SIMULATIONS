Wash- Rest game.
Assume x the frequency of R.
Hence 1-x the frequency of W.

See my notes in the analytic answer.

In order to compute the derivative of x, we use the result from the replicator equation, and hence equals x(1-x)(W(W)-W(R)), meaning:

x(x-1)(b(1-x)-b+k(1+x)/2).

Equilibrium at x=k/(2b-k), and always b>k.

Assume b=4, k=2, and explore what happens.


```
%pylab inline
def WashRest(state,t):

  # unpack the state vector

  x = state

  

  # return the the deriviative of x

  return x*(x-1)*(4*(1-x)-4+2*((1+x)/2))



state0 = 0.5   # start with the population consisting of equal number of wash and rest

t = arange(0.0, 20.0, 0.1)


state = odeint(WashRest, state0, t)



plot(t, state)

xlabel('TIME (sec)')

ylabel('STATES')

title('Wash Rest System')

legend('x');
```
