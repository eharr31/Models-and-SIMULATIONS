```
# I solve part b first
from sympy import *
p, b, k, w0 = symbols("p b k w0")
wr=w0+p*0+(1-p)*b
ww=w0+p*(b-k)+(1-p)*(b-k/2)
wavr=p*wr+(1-p)*ww
deltap=p*(1-p)*(wr-ww)/wavr
solve (deltap,p)
```

I move to part c as in the analytic solution
I examine the conditions in equilibrium 


```
p=k/(2*b-k)
wr=w0+b*(1-k/(2*b-k))
ws=w0+(b-k/2)*(1-(k/(2*b-k)))
sinvadeequi=ws-wr
# assume b=4
sinvadeequi=simplify(sinvadeequi.subs(b,4))
solve(sinvadeequi>0,k)
```

Since b>k, this really means k<0


Wash- Rest game. Assume x the frequency of R. Hence 1-x the frequency of W.

See my notes in the analytic answer.

In order to compute the derivative of x, we use the result from the replicator equation, and hence equals x(1-x)(W(W)-W(R)), meaning:

x(x-1)(b(1-x)-b+k(1+x)/2).

Equilibrium at x=k/(2b-k), and always b>k, from above

Assume b=4, k=2, and explore what happens.



```
#!pip install mpltools
#from mpltools import style
%pylab inline
style.use('ggplot') 
from scipy.integrate import odeint
import numpy as np
def WashRest(state,t):

  # unpack the state vectorz

  x = state

  

  # return the the deriviative of x

  return x*(x-1)*(4*(1-x)-4+2*((1+x)/2))

# solve for different initial states

state0 = 0.2   # start with the population consisting of 0.1R, and 0.9W

t = arange(0.0, 20.0, 0.1)

state1 = odeint(WashRest, state0, t)

state0=0.4

state2 = odeint(WashRest, state0, t)

state0=0.33333333333333333333333

state3 = odeint(WashRest, state0, t)


plt.plot(t,state1,'r-',linewidth=2,label='state0=0.2')

plt.plot(t,state2,'b--',linewidth=2,label='state0=0.4')

plt.plot(t,state3,'g:',linewidth=2,label='state0=0.33')

xlabel('TIME (sec)')

ylabel('STATES')

title('Wash-Rest System')

legend()

plt.show()



# plot dx over x

x = arange(0.0, 1, 0.01)

y = WashRest(x,0)

title('Wash-Rest System, dx over x')

plot(x,y)


axhline(0,color='k')

annotate('k/(2k-b)=0.33', xy=(0.33, 0), xytext=(0.7, 0.01),

            arrowprops=dict(facecolor='black', arrowstyle='simple'),fontsize=16

            );




```
